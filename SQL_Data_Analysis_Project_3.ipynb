```python
import pandas as pd
import sqlite3
```

```python
df_cleaned = pd.read_excel("Cleaned_Dataset_Project1_Final.xlsx")
```

```python
conn = sqlite3.connect(':memory:')
```

```python
df_cleaned.to_sql('ecommerce_sales', conn, index=False, if_exists='replace')
```

```python
# Testing Aggregation Queries
# Function to cleanly format and print SQL query data tables
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
# Test query for product portfolio summaries
product_query = """
SELECT 
    Product,
    COUNT(OrderID) AS Total_Orders,
    ROUND(AVG(TotalPrice), 2) AS Avg_Order_Value,
    ROUND(SUM(TotalPrice), 2) AS Total_Revenue
FROM ecommerce_sales 
GROUP BY Product
ORDER BY Total_Revenue DESC;
"""
```

```python
run_query(product_query)
```

```python
#Uncovering Top Customer Referral Sources
referral_query = """
SELECT 
    ReferralSource,
    COUNT(OrderID) AS Direct_Acquisitions,
    ROUND(SUM(TotalPrice), 2) AS Channels_Attributed_Revenue
FROM ecommerce_sales
GROUP BY ReferralSource
ORDER BY Channels_Attributed_Revenue DESC;
"""
```

```python
run_query(referral_query)
```
